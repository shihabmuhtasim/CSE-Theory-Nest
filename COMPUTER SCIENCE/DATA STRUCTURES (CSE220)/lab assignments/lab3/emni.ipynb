```python
def removeElement(arr, index, size):
  for i in range(index , size):
    arr[i] = arr[i+1] #Shifting left from removing index
  arr[size - 1] = None  #Making last space empty
  return arr
arr=[10,20,30,40,None,None,None,None]
print(removeElement(arr,2,4))
```

```python
def rotateRight(arr):
  temp = arr[len(arr) - 1]
  for i in range(len(arr) - 1, 0, -1):
    arr[i] = arr[i - 1]
  arr[0] = temp
  return arr
arr=[10,20,30,40,None,None,None,None]
rotateRight(arr)
```

```python
def rotateLeft(arr):
  temp = arr[0]
  for i in range(1, len(arr)):
    arr[i-1] = arr[i]
  arr[len(arr) - 1] = temp
  return arr
arr=[10,20,30,40,None,None,None,None]
rotateLeft(arr)
```
