```python
class Node:
    def __init__(self, data, loc):
        self.data = data
        self.next = loc

class MyList:
    def __init__(self, arr):
        if(len(arr) == 0):
            print("Array cannot be empty")
        else:
            self.head = None
            tail = None
            for i in arr:
                n1 = Node(i, None)
                
                if(self.head == None):
                    self.head = n1
                    tail = n1

                else:
                    tail.next = n1
                    tail = n1

    def showList(self):
        n = self.head
        while(n != None):
            if(n.next is not None):
                print(n.data, end="->")
            else:
                print(n.data, end='-|')
            n = n.next
arr = [20, 64, 11, 50, 41, 37]
print("----------1---------")
newList = MyList(arr)
newList.showList()
print()
```
