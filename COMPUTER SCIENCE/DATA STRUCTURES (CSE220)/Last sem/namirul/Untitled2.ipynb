```python
class Node:
    def __init__(self, value, left, right, parent):
        self.value = value
        self.left = left
        self.right = right
        self.parent = parent

def find_max(num1, num2):
    if(num1 > num2):
        return num1
    else:
        return num2

chosen_nodes = [0]*3
    
class Tree:
    def __init__(self):
        self.root = Node(5, None, None, None)
        n1 = Node(1, None, None, None)
        n2 = Node(3, None, None, None)
        n3 = Node(4, n2, None, None)
        n2.parent = n3
        
        n4 = Node(2, n1, n3, self.root)
        n1.parent = n4
        n3.parent = n4
        
        n5 = Node(6, None, None, None)
        n6 = Node(9, None, None, None)
        n7 = Node(8, n5, n6, self.root)
        n5.parent = n7
        n6.parent = n7
        
        self.root.left = n4
        self.root.right = n7
        
        chosen_nodes[0] = n2
        chosen_nodes[1] = n7
        chosen_nodes[2] = n3
    
    def height(self, rt):
        if(rt == None):
            return 0
        else:
            return 1 + find_max(self.height(rt.left), self.height(rt.right))
    
    def level(self, node):
        if(node == self.root):
            return 0
        else:
            return 1 + self.level(node.parent)
        
    def preOrder(self, rt):
        if(rt == None):
            return 0
        else:
            print(rt.value, end=" ")
            self.preOrder(rt.left)
            self.preOrder(rt.right)
            
    def inOrder(self, rt):
        if(rt == None):
            return 0
        else:
            self.inOrder(rt.left)
            print(rt.value, end=" ")
            self.inOrder(rt.right)
    
    def postOrder(self, rt):
        if(rt == None):
            return 0
        else:
            self.postOrder(rt.left)
            self.postOrder(rt.right)
            print(rt.value, end=" ")
    
new_tree = Tree()
print(new_tree.height(new_tree.root))
print(new_tree.level(chosen_nodes[2]))
print()
new_tree.preOrder(new_tree.root)
print()
new_tree.inOrder(new_tree.root)
print()
new_tree.postOrder(new_tree.root)
```
