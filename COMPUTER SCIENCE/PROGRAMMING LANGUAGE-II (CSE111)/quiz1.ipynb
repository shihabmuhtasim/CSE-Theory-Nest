```python
class ShoppingCart:
    def __init__(self):
        self.gadgets = []
        self.cart_name = ""
    def addGadget(self, *gadgets):
        if(self.cart_name == ""):
            print("Please set the cart name first.")
        else:
            for gadget in gadgets:
                self.gadgets.append(gadget)                
    def setCartName(self, cart_name):
        self.cart_name = cart_name        
    def printCartDetails(self):
        print(f'Details of  {self.cart_name}  cart:')
        print(f'Total gadgets in cart: {str(len(self.gadgets))}')
        for gadget in self.gadgets:
            print(gadget)     
    def removeFromCart(self, model):
        flag = False        
        for gadget in self.gadgets:
            if gadget.model == model:
                self.gadgets.remove(gadget)
                flag = True
                break                
        if(flag):
            print(f'{model} removed from cart!')
        else:
            print(f'{model} not found in cart!')
        
class Product:
    totalProducts = 0
    def __init__(self, model, g_name, price, quantity):
        self.model = model
        self.g_name = g_name
        self.price = price
        self.quantity = quantity
        Product.totalProducts = Product.totalProducts + 1
    
    def createProduct(name, product, price, quantity):
        return Product(name, product, price, quantity)
    def __str__(self):
        return "Gadget:" + self.g_name + " Model:" + self.model + " Price:" + str(self.price) + "$ Stock:" + str(self.quantity) + " pcs";
 
s1 = ShoppingCart()
p1 = Product("Razer BlackShark","Headset",99.99,5)
p2 = Product("Razer Huntsman","Keyboard", 249.99,12)
print("1.====================================")
s1.addGadget(p1,p2)
print("2.====================================")
s1.setCartName("Amazon")
s1.addGadget(p1,p2)
s1.addGadget(Product("HyperX Fury", "Mousepad",26.99,21))
print("3.====================================")
s1.printCartDetails()
print("4.====================================")
s1.removeFromCart("Logitech G ProX Superlight")
print("5.====================================")
s1.removeFromCart("Razer Huntsman")
print("6.====================================")
s1.printCartDetails()
```
