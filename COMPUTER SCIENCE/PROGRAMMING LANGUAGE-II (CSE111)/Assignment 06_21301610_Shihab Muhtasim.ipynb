```python
#2
class Assassin:
    count=0
    def __init__(self,name,rate):
        self.name=name
        self.rate=rate
        Assassin.count+=1
        
    def printDetails(self):
        print(f'Name: {self.name}\nSuccess rate: {self.rate}%\nTotal number of Assassin: {Assassin.count}')
    
    @classmethod
    def failurePercentage(cls,name, failure):
        percentage=100-failure
        obj=cls(name,percentage)
        return obj
    @classmethod
    def failureRate(cls,name, failure_rate):
        rate=100-failure_rate
        obj=cls(name,rate)
        return obj
    
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#6
class Laptop:
    laptopCount=0
    def __init__(self,name,count):
        self.name=name
        self.count=count
        Laptop.laptopCount+=count
        
    @staticmethod
    def advantage():
        print("Laptops are portable") 
    @classmethod
    def resetCount(cls):
        cls.laptopCount=0

lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#7
class Cat:
    Number_of_cats=0
    def __init__(self,colour,activity):
        self.colour=colour
        self.act=activity
        Cat.Number_of_cats+=1
    @classmethod
    def no_parameter(cls):
        obj=cls("White","sitting")
        return obj
    @classmethod
    def first_parameter(cls,colour):
        obj=cls(colour,"sitting")
        return obj
    @classmethod
    def second_parameter(cls,activity):
        obj=cls("Grey",activity)
        return obj
    def changeColor(self,new_colour):
        self.colour=new_colour
    def printCat(self):
        print(f'{self.colour} cat is {self.act}')

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#8
import math
class Cylinder:
    radius=5
    height=18
    def __init__(self,radius,height):
        
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
        Cylinder.radius=radius
        Cylinder.height=height
        print(f"Updated: radius={Cylinder.radius} and height={Cylinder.height}.")
        
    @classmethod
    def swap(cls,radius,height):
        obj=cls(height,radius)
        return obj
    @classmethod
    def changeFormat(cls,string):
        radius,height=string.split("-")
        obj=cls(float(radius),float(height))
        return obj
    @staticmethod
    def area(radius,height):
        area=(2*math.pi*(radius)**2)+2*math.pi*radius*height
        print("Area:",area)
    @staticmethod
    def volume(radius,height):
        volume=math.pi*(radius)**2*height
        print("Volume:",volume)

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#10
class SultansDine:
    branch_count=0
    total_sell=0
    list_info=[]
    def __init__(self,name):
        self.name=name
        SultansDine.branch_count+=1
        SultansDine.list_info.append(self.name)
        
    def sellQuantity(self,quantity):
        if quantity < 10:
             self.sell=quantity* 300
        elif 10<= quantity <= 20:
             self.sell=quantity* 350
        else:
            self.sell=quantity* 400  
        SultansDine.total_sell+=self.sell
        SultansDine.list_info.append(self.sell)
        
    def branchInformation(self):
        print(f"Branch Name: {self.name}\nBranch Sell: {self.sell} Taka") 
        
    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {cls.branch_count}\nTotal Sell: {cls.total_sell}")
        if cls.branch_count>0 :
            for index in range(0,len(cls.list_info),2):
                percent=(cls.list_info[index+1]/cls.total_sell)*100
                rate="{:.2f}".format(percent)
                print(f"Branch Name: {cls.list_info[index]}, Branch Sell: {cls.list_info[index+1]} Taka\nBranch consists of total sell's: {rate}%")

SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
