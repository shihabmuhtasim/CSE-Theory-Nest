```python
#1
class Calculator:
    def __init__(self):
        print("Let’s Calculate!")
    def add (self,num1,num2):
        return (num1+num2)
    def substract (self,num1,num2):
        return (num1-num2)
    def multiply (self,num1,num2):
        return num1*num2
    def divide (self,num1,num2):
        return num1/num2
        
val_1= int(input("Enter your value: "))
operator= input("Enter your operator: ")
val_2= int(input("Enter your value: ")) 

obj1=Calculator()
if operator=="+":
    result= obj1.add(val_1,val_2)
elif operator=="-":
    result= obj1.substract(val_1,val_2)
elif operator=="*":
    result= obj1.multiply(val_1,val_2)
elif operator=="/":
    result= obj1.divide(val_1,val_2)
print('Value 1:',val_1)
print('Operator:',operator)
print('Value 2:',val_2)
print("Result:",result)
```

```python
#2
class Customer:
    def __init__(self,name):
        self.name=name
    def greet(self,name_2=None):
        if name_2==None:
            print("Hello!")
        else:
            print(f"Hello,{name_2}!")
    def purchase(self,*items):
        print(f'{self.name}, you purchased {len(items)} item(s):')
        for elem in items:
            print(elem)

customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#5
class Student:
    def __init__(self,name="default"):
        self.name=name
    def quizcalc(self, *marks):
        total=0
        for mark in marks:
            total+=mark
        self.mark=(total/3)
    def printdetail(self):
        if self.name== "default":
            print(f'Hello {self.name} student')
        else:
            print(f'Hello {self.name}')
        print(f'Your average quiz score is {self.mark}')
s1 = Student()
s1.quizcalc(10)
print('--------------------------------')
s1.printdetail()
s2 = Student('Harry')
s2.quizcalc(10,8)
print('--------------------------------')
s2.printdetail()
s3 = Student('Hermione')
s3.quizcalc(10,9,10)
print('--------------------------------')
s3.printdetail()
```

```python
#6
class Vehicle:
    def __init__(self,val_x=0,val_y=0):
        self.val_x=val_x
        self.val_y=val_y
    def moveUp(self):
        self.val_y=self.val_y+1
    def moveLeft(self):
        self.val_x=self.val_x-1
    def moveDown(self):
        self.val_y=self.val_y-1
    def moveRight(self):
        self.val_x=self.val_x+1
    def print_position(self):
        print((self.val_x,self.val_y))
        
car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
#12
class ParcelKoro:
    def __init__(self,name='No name set',weight=0):
        self.name=name
        self.product_weight=weight
    def calculateFee(self,location=None):
        if self.product_weight==0:
            charge=0
        else:
            if location==None:
                location_charge=50
            else:
                location_charge=100
            charge=self.product_weight*20+location_charge
        self.charge=charge
    def printDetails(self):
        print('Customer Name:',self.name)
        print('Product Weight:',self.product_weight)
        print('Total fee:',self.charge)


print("**********************")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.product_weight = 15
p2.calculateFee()
p2.printDetails()
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
#13
class Batsman:
    def __init__(self,*info):
        if len(info)==2:
            self.name='New Batsman'
            self.score=info[0]
            self.ball=info[1]
        elif len(info)==3:
            self.name=info[0]
            self.score=info[1]
            self.ball=info[2]
            
    def setName(self,name):
        self.name=name
        
    def  printCareerStatistics(self):
        print("Name:",self.name)
        print("Runs Scored:",self.score,end=", ")
        print("Balls Faced:",self.ball) 
    def battingStrikeRate(self):
        strike_rate=(self.score/self.ball)*100
        return strike_rate 
        
b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#15
class Account:
    def __init__(self,name='Default Account',balance=0.0):
        self.name=name
        self.balance=balance

    def details(self):
        return f'{self.name}\n{self.balance}'
    
    def withdraw(self,amount):
        new_balance=self.balance-amount
        if new_balance<=3070:
            print("Sorry, Withdraw unsuccessful!The account\nbalance after deducting withdraw amount is\nequal to or less than minimum.")
        else:
            self.balance=new_balance
            print("Withdraw successful! New balance is:",self.balance)

a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930);
print("------------------------")
a2.withdraw(600);
print("------------------------")
a1.withdraw(6929)
```

```python
#19
class Hotel:
    def __init__(self,p_name,g_count=0,s_count=0,dic={}):
        self.p_name=p_name
        self.g_count=g_count
        self.s_count=s_count
        self.dic=dic
    def addStuff(self,name,age,phone_num='000'):
        self.name=name
        self.age=age
        self.phone_num=phone_num
        print(f'Staff With ID {self.s_count+1} is added')
        self.s_count+=1
        
    def getStuffById(self,id_no):
        self.id=id_no
        return f'Staff ID: {self.id}\nName: {self.name}\nAge: {self.age}\nPhone no.: {self.phone_num}'
         
    def addGuest(self,g_name,g_age,g_phone_num):    
        print(f'Guest With ID {self.g_count+1} is created')
        self.g_name=g_name
        self.g_age=g_age
        self.g_phone_num=g_phone_num
        self.dic[self.g_count+1]=[self.g_name,self.g_age,self.g_phone_num]
        self.g_count+=1
        print(self.dic)
       

    def getGuestById(self,g_id):
        self.g_id=g_id
        return f'Staff ID: {self.g_id}\nName: {self.g_name}\nAge: {self.g_age}\nPhone no.: {self.g_phone_num}'
        
    def allStaffs(self):
        print('All Staff:')
        print(f'Number of staff:{self.s_count}')
        print(f'Staff ID: {self.id} Name: {self.name} Age: {self.g_age} Phone no.: {self.phone_num}')
    def allGuest(self):
        print(f'All Guest:')
        print(f'Number of guest:{self.g_count}')
        for key,val in self.dic.items():
                print(f'Staff ID: {key} Name: {val[0]} Age: {val[1]} Phone no.: {val[2]}')
            
        
        

h = Hotel("Lakeshore")
h.addStuff( "Adam", 26)
print("=================================")
print(h.getStuffById(1))
print("=================================")
h.addGuest('Carol',35,'123')
print("=================================")
print(h.getGuestById(1))
print("=================================")
h.addGuest("Diana", 32, '431')
print("=================================")
print(h.getGuestById(2))
print("=================================")
h.allStaffs()
print("=================================")
h.allGuest()
```
